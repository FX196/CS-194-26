```python
import numpy as np
import skimage as sk
import skimage.io as skio
from scipy.signal import convolve2d, unit_impulse
import cv2
from matplotlib import pyplot as plt
from scipy.ndimage.interpolation import rotate
from scipy.ndimage import gaussian_filter
from tqdm import tqdm
from align_image_code import align_images
# plt.use('TkAgg')
```

```python
data_dir = "data/"
imname = "cameraman.png"

def imshow(img):
    skio.imshow(img)
    skio.show()
```

```python
def g(size, sigma):
    gkern = cv2.getGaussianKernel(size, sigma)
    return gkern @ gkern.T
```

```python
def color_conv(im, kern):
    res = np.zeros(im.shape)
    for i in range(3):
        res[:, :, i] = convolve2d(im[:, :, i], kern, mode="same")
    return res
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
im = skio.imread(data_dir + imname, as_gray=True)
im = sk.img_as_float(im)
```

```python
panda = skio.imread(data_dir+"kwason.jpg")/255
skio.imshow(panda)
```

```python
blur_panda = color_conv(panda, g(12, 2))
skio.imshow(blur_panda)
```

```python
def sharpened(im, size, sigma, alpha):
    gkern = cv2.getGaussianKernel(size, sigma)
    gfilter = gkern @ gkern.T
    im_low = np.zeros(im.shape)
    for i in range(3):
        im_low[:, :, i] = convolve2d(im[:, :, i], gfilter, mode="same")
    return im+alpha*(im-im_low)
```

```python
skio.imshow(sharpened(blur_panda, 12, 2, 2))
```

```python
lincoln = plt.imread("data/lincoln.jpg")/255
```

```python
def gaussian_stack(im, layers=3, size=15, sigma=8):
    g_stack = [im]
    for i in range(layers):
        g_stack.append(color_conv(g_stack[i], g(size, sigma)))
    return g_stack
```

```python
for frame in gaussian_stack(lincoln):
    plt.figure()
    plt.imshow(frame)
```

```python
def laplacian_stack(im, layers=3, size=15, sigma=8):
    g_stack = gaussian_stack(im, layers, size, sigma)
    l_stack = []
    for i in range(len(g_stack)-1):
        l_stack.append(g_stack[i+1] - g_stack[i])
    l_stack.append(g_stack[-1])
    return l_stack
```

```python
for frame in laplacian_stack(lincoln):
    plt.figure()
    plt.imshow(frame/np.max(frame))
```
